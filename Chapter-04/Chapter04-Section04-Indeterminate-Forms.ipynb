### Chapter 4, Section 4: Indeterminate Forms and l'Hospital's Rule
- Problems: Paper (7, 9, 15, 21, 57)
- Problems: Jupyter (71).

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    x = np.asarray(x, dtype=float)
    return np.exp(x * np.log1p(2.0 / x))

# x values: small integers and a spread up to 10000
x_vals = np.concatenate((np.arange(1,51), np.linspace(100, 10000, 200)))
x_vals = np.unique(x_vals)
y_vals = f(x_vals)

e2 = np.exp(2.0)

plt.figure(figsize=(8,4))
plt.plot(x_vals, y_vals, marker='o', markersize=3, linestyle='-', label='(1+2/x)^x')
plt.axhline(e2, color='red', linestyle='--', label='e^2')
plt.xscale('log')
plt.xlabel('x (log scale)')
plt.ylabel('Value')
plt.title('Convergence of (1 + 2/x)^x to e^2')
plt.legend()
plt.tight_layout()
plt.show()

# quick numeric check
print("x (last) =", int(x_vals[-1]), "value =", y_vals[-1])
print("e^2 =", e2)

```
