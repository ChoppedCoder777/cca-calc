```python
# Problem 31, page 287
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 397)
fx = 3*x**4 + 8*x**3 - 48*x**2
fprime = 12*x**3 + 24*x**2 - 96*x
plt.plot(x, fx, label=r'$3x^4 + 8x^3 - 48x^2$', color='blue')
plt.plot(x, fprime, label=r'$12x^4 + 24x^2 - 96x$', color='red')
plt.title('')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Compute the derivative of fx symbolically using sympy
import sympy as sp
x = sp.symbols('x')
fx_sym = 3*x**4 + 8*x**3 - 48*x**2
fprime_sym = sp.diff(fx_sym, x)
print('Symbolic derivative of fx:')
print(fprime_sym)

# Convert symbolic expressions to numerical functions
fx_num = sp.lambdify(x, fx_sym, modules=['numpy'])
fprime_num = sp.lambdify(x, fprime_sym, modules=['numpy'])

# Generate x values and evaluate functions
x0 = np.linspace(-5, 5, 400)
y_fx = fx_num(x0)
y_fprime = fprime_num(x0)

# Plot the functions
plt.plot(x0, y_fx, label=r'$3x^4 + 8x^3 - 48x^2$', color='blue')
plt.plot(x0, y_fprime, label=r'Derivative', color='red')
plt.title('Function and Derivative')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Compute the derivative of fx symbolically using sympy
import sympy as sp
x = sp.symbols('x')
fx_sym = x**(-2)*sp.log(x)
fprime_sym = sp.diff(fx_sym, x)
print('Symbolic derivative of fx:')
print(fprime_sym)

# Convert symbolic expressions to numerical functions
fx_num = sp.lambdify(x, fx_sym, modules=['numpy'])
fprime_num = sp.lambdify(x, fprime_sym, modules=['numpy'])

# Generate x values and evaluate functions
x0 = np.linspace(1.5, 1.75, 400)
y_fx = fx_num(x0)
y_fprime = fprime_num(x0) 
# Find absolute min and max of fx_sym on [1/2, 4]
interval = [sp.Rational(1,2), 4]
# Find critical points in the interval
critical_points = sp.solve(fprime_sym, x)
critical_points_in_interval = [pt.evalf() for pt in critical_points if pt.is_real and interval[0] <= pt.evalf() <= interval[1]]
# Evaluate at endpoints and critical points
test_points = [interval[0], interval[1]] + critical_points_in_interval
values = [(pt, fx_sym.subs(x, pt).evalf()) for pt in test_points]
min_pt, min_val = min(values, key=lambda t: t[1])
max_pt, max_val = max(values, key=lambda t: t[1])
print(f"\nAbsolute minimum on [1/2, 4]: x = {min_pt}, f(x) = {min_val}")
print(f"Absolute maximum on [1/2, 4]: x = {max_pt}, f(x) = {max_val}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3*x**2 + x - 2

# Create x values
x = np.linspace(-10, 10, 400)

# Compute y values
y = f(x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x) = 3x + x - 2", color="blue")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.title("Graph of f(x) = 3x + x - 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(y):
    return (y - 1) / (y**2 - y + 1)

# Create y values
y = np.linspace(-10, 10, 400)

# Compute f(y)
fy = f(y)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(y, fy, label="f(y) = (y - 1) / (y - y + 1)", color="purple")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.title("Graph of f(y) = (y - 1) / (y - y + 1)")
plt.xlabel("y")
plt.ylabel("f(y)")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(t):
    return 2*np.cos(t) + np.sin(2*t)

# Create t values on the interval [0, /2]
t = np.linspace(0, np.pi/2, 400)

# Compute y values
y = f(t)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(t, y, label="f(t) = 2 cos(t) + sin(2t)", color="orange")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.title("Graph of f(t) = 2 cos(t) + sin(2t) on [0, /2]")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x * np.sqrt(x - x**2)

# Domain: where x - x^2 >= 0    [0,1]
x = np.linspace(0, 1, 400)

# Compute y values
y = f(x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r"$f(x) = x\sqrt{x - x^2}$", color="blue")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.title("Graph of f(x) = x(x - x) on [0, 1]")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

# Print approximate max/min values
print("Approximate minimum value:", np.min(y))
print("Approximate maximum value:", np.max(y))

```
